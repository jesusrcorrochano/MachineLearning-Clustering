```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster import hierarchy
```

```python
dataset = pd.read_csv('gender_voice_dataset_normalizado_sinlabel.csv')
```

```python
dataset
```

```python
dataset=dataset.sample(150)
```

```python
dataset
```

```python
X = dataset.values
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(X, method  = "ward"))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
Z = linkage(X, 'ward')
```

```python
last = Z[-10:, 2]
last_rev = last[::-1]
idxs = np.arange(1, len(last) + 1)
plt.plot(idxs, last_rev)

acceleration = np.diff(last, 2)  # 2nd derivative of the distances
acceleration_rev = acceleration[::-1]
plt.plot(np.flip(idxs), last, marker='o', markerfacecolor='green', markersize=10, color='lightgreen', linewidth=3)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Euclidean distance")
plt.show()
k = acceleration_rev.argmax() + 2  # if idx 0 is the max of this we want 2 clusters
print("clusters:", k)
```

```python
plt.plot(np.flip(idxs), last)
plt.show()
```

```python
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
hierarchy.set_link_color_palette(['green', 'red', 'blue', 'gold'])
hierarchy.dendrogram(
    Z,
    color_threshold=4, above_threshold_color='black',
    leaf_rotation=90.,  # rotates the x axis labels
    leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```
